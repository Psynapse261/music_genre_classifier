```python
#Importing libraries
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
#Importing the dataset
dataset = pd.read_csv('Data/features_30_sec.csv')
print('No. of rows: ' + str(len(dataset)))
dataset.head()
```

```python
#Splitting the data
X = dataset.iloc[:,2:59]
Y = dataset.iloc[:,59]
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

```python
# Scaling all the features
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
```

```python
# Getting the value of K
import math
math.sqrt(len(x_train))
```

```python
# Making the classifier and getting predictions
knn_classifier = KNeighborsClassifier(n_neighbors=27, p = 2, metric='euclidean')
knn_classifier.fit(x_train, y_train)

#Getting predictions on test data
y_pred = knn_classifier.predict(x_test)
```

```python
# Getting the model's confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Evaluating the F1 score
print('F1 score: ' + str(f1_score(y_test, y_pred)))
```

```python
#Displaying the accuracy
print('Accuracy: ' + str(accuracy_score(y_test, y_pred)))
```
